## Part 1 : Text cleaning using re / regex

A RegEx (Regular Expression), is a sequence of characters that forms a search pattern.

RegEx can be used to check if a string contains the specified search pattern.

## Applications:
1.   Remove unwanted characters with sub(pattern, replacement, string)

```python
import re
text = "Contact info: (123)-456-7890 and 987-654-3210."
cleaned_text = re.sub(r'[()-]', '', text)
print(cleaned_text)
```

2.   Extract specific patterns with findall(pattern, string)



```python
text = "Please reach out to us at support@example.org or help@example.org."
emails = re.findall(r'\b[\w.-]+?@\w+?\.\w+?\b', text)
print(emails)
```

3. Validate data formats with match()

```python
email = "test@example.com"
if re.match(r'^\b[\w.-]+?@\w+?\.\w+?\b$', email):
    print("Valid email")
else:
    print("Invalid email")
```

4. Split strings by patterns with split(pattern, string)

```python
text = "This is sentence one. And this is sentence two! Is this sentence three?"
sentences = re.split(r'[.!?]', text)
print(sentences)
```

## Metacharacters:
Are used to specify the search. Some examples:


*   [] - a set of characters - "[a-m]"
*   \ - signals a special sequence - "\d"
*   . - any character - "he..o"
*   ^ - starts with - "^hello"

--> you can combine regex with pandas for efficient cleaning of text fields in data frames.

## Part 3 : Normalization
Source: https://www.ibm.com/think/topics/stemming-lemmatization

###Stemming
Stemming uses a set of basic rules to chop off the ends of words (suffixes) in hopes of finding the root. It’s fast and efficient, but it doesn't care about grammar or vocabulary.


```python
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

# Initialize the stemmer
ps = PorterStemmer()

text = "Love looks not with the eyes but with the mind, and therefore is winged Cupid painted blind."
words = word_tokenize(text)

# Stem the words
stemmed_words = [ps.stem(word) for word in words]

print(stemmed_words)
```

###Lemmatization
Lemmatization looks at the word's context and consults a dictionary (a lexicon) to return the lemma, which is the actual dictionary base form of the word.

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
from nltk import word_tokenize, pos_tag

def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN

def lemmatize_passage(text):
    words = word_tokenize(text)
    pos_tags = pos_tag(words)
    lemmatizer = WordNetLemmatizer()
    lemmatized_words = [lemmatizer.lemmatize(word, get_wordnet_pos(tag)) for word, tag in pos_tags]
    lemmatized_sentence = ' '.join(lemmatized_words)
    return lemmatized_sentence

text = "There is nothing either good or bad but thinking makes it so."
result = lemmatize_passage(text)

print("Original:", text)
print("Tokenized:", word_tokenize(text))
print("Lemmatized:", result)
```

## Part 4 : Bag-of-Words model


The bag of words model is a text representation technique used in Natural Language Processing (NLP).
Text is converted to numerical vecors by counting how many times they appear in the text.


## Application example of BoW : Spam detection

1.   Email 1: Win money now
2.   Email 2: Meeting tomorrow

#### After Bag-of-Words:
Vocabulary vector
| Email | win | money | now | meeting | tomorrow |
|------|-----|-------|-----|--------|----------|
| Email 1 | 1 | 1 | 1 | 0 | 0 |
| Email 2 | 0 | 0 | 0 | 1 | 1 |

These vectors go into a classifier like:
* Logistic Regression
* Naive Bayes
* SVM

### The model learns patterns:
Groups of words like win, money, free → **spam**.


```python
#Examples BoW

from sklearn.feature_extraction.text import CountVectorizer

corpus = [
    "Congratulations! You have won $10,000 in our lottery. Claim your prize now.",
    "Earn money fast from home. No experience required. Start making $500 per day today.",
    "Limited offer! Invest $100 today and double your money in just one week"
]

count = CountVectorizer()
X = count.fit_transform(corpus)

print(count.get_feature_names_out())
print(X.toarray())

```

## Part 4 : TF–IDF

Stands for *Term-Frequency Inverse Document Frequency* and is part of Natural Language Processing (NLP).
TF-IDF emphasizes important words that appear frequently in a document but appear rarely in the corpus (i.e. a collection of documents). It is what sets this document appart from others and thus hints to the nature or specialty of the document.
* Not stop words (very frequent words across *all* documents)
* The words that appear many times in a document but are not stop words -> special

* E.g., a document that mentions the word “medicine” 10 times, it's most likely about that topic
* E.g., a document mentioning the word “Achalasia”, a rare disorder of the esophagus

**Term Frequency (TF)** is defined as the number of times a term *t* in document *d*, as ratio of the term to the total number of terms in that document.
The TF formula is given by:

**Inverse Document Frequency (IDF)** measures the importance of term *t* in the entire corpus *D*.

A high TF-IDF is reached by a high term frequency in a given document, and a low document frequency of the term in all documents.

$$\text{TF}(t,d) = \frac{\text{No. of term t in document d}}{\text{Total no. of terms in document d}}$$

$$\text{IDF}(t,D) = \log \left(\frac{\text{Total no. of documents in corpus D}}{\text{No. of documents containing term t}}\right)$$

$$\text{TF-IDF}(t,d,D) = \text{TF}(t,d) \times \text{IDF}(t,D)$$

 The goal of IDF is to penalize words that are common across all documents.

TF-IDF is helpful to determine individual words' importance. The metric is effective on smaller and larger data sets, giving you flexibility in applying it.

Limitations:
* As a mathematical calculation, the program can’t understand that a word is the same when used in different tenses or with different forms. It would classify “create,” “created,” “creates,” and “creating” as four different words when, for most analytical purposes, they are the same.
* Another problem lies in compound nouns like the “White House.” TF-IDF doesn’t have a mechanism to understand that these words are related.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

corpus = ["good dog", "good cat", "good dog cat"]
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(corpus)
print(X.toarray())

'''
Inside this array of array the first number represents cat, the second number
represents dog and the third number represents good (alphabetically Cat -> Dog -> Good).

The length of the words in the corpus is 3 and they are arranged in ascending order.
'''
```

## Main Takeaways

### 1. Text needs preprocessing
Raw text cannot be directly used by machine learning models.  
Steps like **cleaning, tokenization, stemming, and lemmatization** help prepare the text for analysis.

### 2. Regular Expressions (RegEx)
RegEx helps clean and extract patterns from text:
- Remove unwanted characters
- Extract emails or phone numbers
- Validate formats
- Split text into parts

### 3. Tokenization
Tokenization splits text into **individual words (tokens)**.  
This is one of the first steps in NLP because models work with tokens rather than full sentences.

### 4. Normalization
Normalization simplifies words to their base form:
- **Stemming:** cuts word endings (fast but less precise)
- **Lemmatization:** returns the correct dictionary form (more accurate)

### 5. Text Encoding
Text must be converted into **numerical vectors** so machine learning models can process it.

Two common techniques:
- **Bag-of-Words (BoW):** counts how often each word appears
- **TF-IDF:** highlights important words that are frequent in a document but rare across all documents

### 6. Key Idea of NLP Pipelines
A typical NLP workflow looks like:

Text → Cleaning → Tokenization → Normalization → Vectorization → Machine Learning Model

These steps allow computers to analyze and learn patterns from human language.
