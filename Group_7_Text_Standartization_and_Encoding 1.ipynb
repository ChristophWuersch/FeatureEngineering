# 1. Historical view: word vectors


### The basic problem: words vs numbers

Machine learning models cannot work directly with text.  
They require **numerical inputs**, so words must first be converted into numbers.

One of the earliest approaches was **one-hot encoding**, where each word is represented by a vector containing mostly zeros and a single one.

Example vocabulary:

| Word | Vector |
|------|--------|
| cat  | [1, 0, 0] |
| dog  | [0, 1, 0] |
| mat  | [0, 0, 1] |

This representation is simple but has a major limitation:  
it does **not capture meaning**.

For example, the distance between **cat** and **dog** is exactly the same as the distance between **cat** and **banana**, even though their meanings are clearly different.

This limitation motivated researchers to search for representations that could capture **semantic relationships between words**.

---

### The Distributional Hypothesis

A foundational idea from linguistics is the **distributional hypothesis**.

In the 1950s, **Zellig Harris** and **J. R. Firth** proposed that:

> Words that appear in similar contexts tend to have similar meanings.

Firth summarized this idea with the now-famous phrase:

> "You shall know a word by the company it keeps."

Consider the word **bank**. Its meaning shifts entirely depending on the surrounding context:

- *River, fish and flood* → **bank** refers to the side of a river
- *Loan, interest and deposit* → **bank** refers to a financial institution

The word is identical in both cases, but the context reveals its meaning. This is precisely what the distributional hypothesis captures.

A contradicting idea focused on grammar and syntax as the core of language — essentially saying meaning comes from formal rules.

---

## Co-occurrence and vector space models

To operationalize this idea, researchers started counting **how often words appear near each other** in large text corpora.

This produced **word–context co-occurrence matrices**, where each row describes the contexts in which a word appears.

Example:

| Word | sits | barks | mat |
|------|------|-------|-----|
| dog  | 3    | 8     | 1   |
| cat  | 4    | 0     | 6   |

Words with similar rows are assumed to have similar meanings.

This approach introduced the idea of representing words as **vectors in a high-dimensional space**.

---

## From sparse matrices to dense representations

However, these co-occurrence matrices quickly became **very large and sparse**, making them inefficient to use directly.

To solve this problem, researchers introduced **dimensionality reduction methods** such as **Latent Semantic Analysis (LSA)**.

LSA compresses the large matrix into a smaller vector space while preserving important semantic relationships.

This step was important because it showed that meaning could be captured using **dense numerical vectors instead of large count tables**.

---

### Word embeddings

Modern **word embeddings** represent each word as a dense vector of real numbers learned from large text corpora.

Example:

- cat → [0.21, -0.14, 0.73]  
- dog → [0.19, -0.10, 0.71]

Now similar words have similar vectors in the embedding space.

One famous example illustrating this property is:

**king − man + woman ≈ queen**

This shows that word vectors can capture **semantic relationships between words**.

![description](https://assets.zilliz.com/Figure_Word_Embeddings_b021a5a759.png)

---

### Historical importance

Word vectors represent a bridge between several traditions:

- linguistic theory (distributional semantics)
- statistical text analysis
- modern neural networks

They laid the foundation for later techniques such as **Word2Vec**, **GloVe**, and the contextual embeddings used in modern **large language models (LLMs)**.

---

## Takeaways

- Machine learning requires text to be converted into numerical representations.
- One-hot encoding was an early solution but does not capture meaning.
- The **distributional hypothesis** introduced the idea that context reveals meaning.
- Co-occurrence matrices were the first practical implementation of this idea.
- Dimensionality reduction methods like **LSA** produced dense semantic vectors.
- Modern **word embeddings** build on these ideas and form the basis of modern NLP.

# 2. Word2Vec

Word2Vec is a method for learning numerical vector representations (embeddings) of words from text data.
Instead of representing words as simple indices or one-hot vectors, Word2Vec trains a small neural network to predict relationships between words in a sentence.
During training, the model learns word vectors such that words appearing in similar contexts obtain similar vector representations.
These vectors capture semantic relationships between words and can be used as features in machine learning models.

The idea is based on the distributional hypothesis: words that occur in similar contexts tend to have similar meanings.

````python
!pip install gensim
````

````python
from gensim.models import Word2Vec
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('punkt_tab')
````

````python
sentences_raw = [
    "the cat sits on the mat",
    "the dog sits on the floor",
    "the dog chased the cat",
    "cats and dogs are animals",
    "the cat climbed the tree"
]

sentences = [word_tokenize(s.lower()) for s in sentences_raw]

sentences
````

## 2.1 Skip-gram

Skip-gram is a Word2Vec training method that predicts surrounding context words from a given center word.
For each word in a sentence, the model tries to predict the words that appear within a defined window around it.
By repeatedly performing this prediction task, the neural network learns vector representations (embeddings) in which words that occur in similar contexts have similar vectors.
Skip-gram typically performs well for rare words because each word generates multiple training examples.

Example sentence:

the cat sits on the mat

Training pairs with window size = 2:

Input: sits → Output: cat  
Input: sits → Output: on

````python
#Modell training based on the Sentences defined in Word2Vec above
model = Word2Vec(
    sentences,
    vector_size=50,   # dimension of word vectors
    window=2,         # context window size
    min_count=1,      # include rare words
    sg=1              # 1 = Skip-gram, 0 = CBOW
)
````

````python
# Vector of a word
model.wv["cat"]
````

````python
# How similar is a word to others
model.wv.most_similar("cat")
#cat
#dog
#climbed
````

````python
# The Vocubular of the Model based on its frequency
model.wv.index_to_key
````

## 2.2 CBOW (Continuous Bag of Words)

CBOW is a **Word2Vec training method** that learns word embeddings by **predicting a target word from its surrounding context words**.

Example sentence:

```
The quick brown fox
```

CBOW training sample:

```python
["The", "quick", "fox"] -> "brown"
```

Interpretation:

```
"The quick ___ fox"
```

The model learns that the missing word is likely **brown**.

---

# Code Example 1 — Generate CBOW training samples

```python
text = "the quick brown fox jumps over the lazy dog"
tokens = text.split()

window = 2
samples = []

for i in range(window, len(tokens)-window):
    context = tokens[i-window:i] + tokens[i+1:i+window+1]
    target = tokens[i]
    samples.append((context, target))

samples
```

Example output:

```
(['the','quick','fox','jumps'], 'brown')
```

# 3. Byte-pair Encoding

**Byte-Pair Encoding (BPE)** was originally introduced in 1994 by Philip Gage as a straightforward and highly effective data compression technique. Long before its adaptation for modern AI and text tokenization, it was designed simply to reduce the file size of digital text and binary data by finding and replacing repeating patterns.


### Core Mechanism

The original algorithm operates through a greedy, iterative process of substitution. Here is how it works:

1. **Frequency Analysis:** The algorithm scans the entire dataset to identify the most frequently occurring pair of adjacent bytes.
2. **Substitution:** It replaces every instance of that frequent pair with a single, **unused byte** that does not currently appear anywhere in the data.
3. **Table Generation:** It records this substitution in a lookup table or dictionary (for example, documenting that the newly introduced byte `Z` represents the original pair `A` and `B`).
4. **Iteration:** The process repeats—scanning the newly compressed data for the next most frequent pair (which can include the newly substituted bytes)—until either no pairs appear more than once or all 256 possible byte values are in use.

To restore the original data, a decompression algorithm simply reads the compressed file and uses the generated lookup table in reverse, recursively expanding the substituted bytes back into their original pairs.

## Final Result Summary

| Step | Merge Rule | New Symbol | String After Merge |
|------|-----------|------------|-------------------|
| 0 | — | — | `aaabdaaabacbdaabafefgef` |
| 1 | `aa → Z` | Z | `ZabdZabacbdZbafefgef` |
| 2 | `Za → Y` | Y | `YbdYbacbdZbafefgef` |
| 3 | `Yb → X` | X | `XdXacbdZbafefgef` |
| 4 | `fe → W` | W | `XdXacbdZbaWfgef` |
| 5 | — (stop) | — | No pair with frequency > 1 |

### Final Compressed String

```
X d X a c b d Z b a W f g e f
```

### Final Vocabulary (Merge Table)

```
Z  = aa
Y  = Za  = aaa
X  = Yb  = aaab
W  = fe
```

### Decompression (Reverse)

To reconstruct the original string, apply merges in **reverse order**:

```
XdXacbdZbaWfgef
→ (W→fe)  XdXacbdZbafefgef
→ (X→Yb)  YbdYbacbdZbafefgef
→ (Y→Za)  ZabdZabacbdZbafefgef
→ (Z→aa)  aaabdaaabacbdaabafefgef  ✓
```

---

### Key Observations

1. **Greedy left-to-right:** When `aaa` is encountered, the first two `a`s merge into `Z`, leaving one `a` behind. The algorithm does **not** look ahead.

2. **Hierarchical merges build up:** Each new symbol can be part of the next merge — `aa→Z`, then `Za→Y`, then `Yb→X`. This builds a **hierarchy of subword units**: `X` = `aaab` was discovered bottom-up.

3. **Frequency-driven:** The algorithm always targets the most common pair, achieving maximum compression per step.

4. **In NLP context:** Instead of single characters, the initial "characters" are bytes or unicode characters, and instead of arbitrary letters like Z/Y/X, the merged tokens get integer IDs. The merge table is saved and reused at inference time to tokenize new, unseen text.

# 4. Modern encoding using LLMs

Large Language Models (LLMs) process language by converting text into numerical representations.
Since computers cannot directly understand human language, text must first be encoded into numbers.
This encoding process enables AI systems to analyze and interpret language.
Modern encoding techniques capture not only words but also their meaning and context.

### The Problem with Word2Vec

Modern encoding techniques rely on the Transformer architecture.
Transformers use a mechanism called self-attention.
Self-attention allows each word in a sentence to consider the other words around it.
This helps the model understand relationships and dependencies between words.

Word2Vec gives every word a **single fixed vector** — "bank" always maps to the same point in space, whether you mean a riverbank or a financial institution. The word's meaning is frozen at training time.

Modern LLMs solve this with **contextual embeddings**: the vector for a word changes depending on the sentence it appears in.

---

### The Core Idea

A transformer-based LLM reads the entire input sequence at once and produces one embedding vector **per token**, where each vector encodes not just the word but its role and meaning in that specific context.
```
"I deposited money at the bank."   → bank vector ≈ [finance region of space]
"We picnicked by the river bank."  → bank vector ≈ [nature region of space]
```

Same word, completely different vectors. This is the fundamental leap over Word2Vec.

### How It Works
```
Input text
    ↓
Tokenize  (split into subword tokens via BPE — see part 3. BPE)
    ↓
Token embeddings  (lookup table, like Word2Vec — still static at this stage)
    ↓
Transformer layers  (self-attention lets every token look at every other token)
    ↓
Contextual output embeddings  (one vector per token, now context-aware)
```

The transformer layers are where the magic happens. Self-attention allows "bank" to look at "river" or "money" in the same sentence and shift its representation accordingly.

````python
# 1. Load Model and Tokenizer
from transformers import AutoTokenizer, AutoModel
import torch

model_name = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
model.eval()
````

````python
# Tokenize the inputs

sentences = [
    "I deposited money at the bank.",
    "We picnicked by the river bank."
]

for sentence in sentences:
    tokens = tokenizer.tokenize(sentence)
    print(f"Sentence : {sentence}")
    print(f"Tokens   : {tokens}")
    print(f"Count    : {len(tokens)} tokens\n")
````

````python
# Extract Token-Level Embeddings

for sentence in sentences:
    inputs = tokenizer(sentence, return_tensors="pt")
    with torch.no_grad():
        outputs = model(**inputs)

    # last_hidden_state shape: [batch_size, num_tokens, hidden_size]
    token_embeddings = outputs.last_hidden_state[0]

    print(f"Sentence : {sentence}")
    print(f"Shape    : {token_embeddings.shape}  → (num_tokens, 768)")
    print(f"          = {token_embeddings.shape[0]} tokens, each a 768-dim vector\n")
````

````python
# Extract the "bank" Vector from Each Sentence

def get_token_embedding(sentence, target_word, tokenizer, model):
    inputs = tokenizer(sentence, return_tensors="pt")
    tokens = tokenizer.convert_ids_to_tokens(inputs["input_ids"][0])

    with torch.no_grad():
        outputs = model(**inputs)

    token_embeddings = outputs.last_hidden_state[0]

    # Find the position of our target word
    token_idx = tokens.index(target_word)
    print(f"Tokens        : {tokens}")
    print(f"'{target_word}' found at position {token_idx}")
    return token_embeddings[token_idx]

bank_finance = get_token_embedding(sentences[0], "bank", tokenizer, model)
print()
bank_nature  = get_token_embedding(sentences[1], "bank", tokenizer, model)
````

````python
# Compare the Two "bank" Vectors

from torch.nn.functional import cosine_similarity

sim = cosine_similarity(bank_finance.unsqueeze(0), bank_nature.unsqueeze(0)).item()
print(f"Cosine similarity between 'bank' (finance) and 'bank' (nature): {sim:.4f}")

# For reference: what do similarity scores mean?
print("\nReference scale:")
print("  1.00 = identical vectors (same word, same context)")
print("  0.90+ = very similar meaning")
print("  0.70  = related but distinct")
print("  <0.50 = clearly different meanings")
````

Interpretation: the two "bank" vectors are moderately similar — they share surface-level word identity — but BERT has pushed them apart in embedding space because their surrounding context words ("money" vs "river") signal different meanings.

Compare this to Word2Vec: it would give cosine similarity = 1.00 for the same word, because Word2Vec has only ONE vector per word regardless of context.

````python
# Sentence-Level Embedding via Mean Pooling

def get_sentence_embedding(sentence, tokenizer, model):
    inputs = tokenizer(sentence, return_tensors="pt")
    with torch.no_grad():
        outputs = model(**inputs)
    # Average all token vectors → one vector for the whole sentence
    return outputs.last_hidden_state.mean(dim=1).squeeze()

emb1 = get_sentence_embedding(sentences[0], tokenizer, model)
emb2 = get_sentence_embedding(sentences[1], tokenizer, model)

sim = cosine_similarity(emb1.unsqueeze(0), emb2.unsqueeze(0)).item()
print(f"Sentence embedding shape : {emb1.shape}  → one 768-dim vector per sentence")
print(f"Sentence-level similarity: {sim:.4f}")
````

The sentences are more similar at the sentence level than at the word level — they share structure, topic (bank), and grammar. Mean pooling blends all token signals together, smoothing out the word-level difference we saw in Step 5.

This is a common trade-off: sentence embeddings lose token-level precision but are practical for tasks like semantic search or classification.

### Word2Vec vs. Modern LLM Encodings

| Property | Word2Vec | LLM (e.g. BERT) |
|---|---|---|
| Embedding per word | One fixed vector | Different vector per context |
| Handles polysemy | ✗ | ✓ |
| Captures syntax | Weakly | Strongly (via attention) |
| Requires fine-tuning for tasks | ✓ (usually) | Often works out of the box |
| Computational cost | Low | High |

### Key Takeaways

- **Word2Vec**: one word = one vector, always. "bank" is always the same point in space no matter what sentence it's in.
- **BERT**: one word = different vector depending on context. "bank" near "money" lands in a different region of space than "bank" near "river". Same spelling, different meaning, different vector.
- Every token gets its own 768-number description. That's the embedding. More numbers = more room to encode nuance.
- `[CLS]` and `[SEP]` are special tokens BERT adds automatically — they're not your words, but they take up slots in the output.
- To go from token embeddings → one vector for the whole sentence: just average all token vectors. This is called **mean pooling**.
- **Cosine similarity** tells you how close two vectors are in meaning: 1.0 = identical, 0.0 = unrelated. The two "bank" sentences score ~0.55 — contextually different, which is exactly what we want.
- BERT reads the whole sentence at once (left AND right context). GPT reads left to right only. That's why BERT is preferred for understanding/embeddings and GPT for generating text.

## Overall Takeaways

This notebook traces a single idea from its origins to modern practice: **how do we give a machine a useful numerical representation of a word?**

Each section answers a limitation of the previous one:

**One-hot encoding** (Section 1) was the naive starting point — simple, but completely blind to meaning. "Cat" and "dog" are just as far apart as "cat" and "volcano."

**Word2Vec** (Section 2) was the first real breakthrough. Instead of hand-crafting vectors, the model learns them from data by exploiting context. Skip-gram and CBOW are two strategies for the same goal: words that appear in similar contexts should end up in similar regions of vector space. The result — vectors where `king − man + woman ≈ queen` — showed that meaning has geometry.

**BPE** (Section 3) solved a practical problem that Word2Vec ignored: what do you do with words you have never seen before? By breaking words into reusable subword pieces, BPE ensures the vocabulary stays manageable and unknown words can still be represented. This is now the standard tokenisation strategy in every modern LLM.

**Contextual embeddings with BERT** (Section 4) closed the final gap. Word2Vec gives every word one fixed vector for life — "bank" is always the same point regardless of whether it sits next to "river" or "interest rate." BERT generates a fresh vector for every word in every sentence, shaped by what surrounds it. Same word, different meaning, different vector.

**The through-line:** we went from counting to learning, from whole words to subwords, and from static to contextual. Each step added expressiveness at the cost of complexity. Modern NLP lives at the BERT end of this spectrum — but understanding all four steps is what makes the final result legible rather than magical.
